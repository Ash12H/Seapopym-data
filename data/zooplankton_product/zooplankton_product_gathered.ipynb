```python
import xarray as xr
from datetime import datetime
```

Load the forcing at each station.

Add the latitude and longitude coordinates.


```python
obs_bats = xr.open_dataset("./bats_zooplankton.nc")
obs_bats
```

```python
obs_hot = xr.open_dataset("./hot_zooplankton.nc")
obs_hot
```

```python
obs_papa = xr.open_dataset("./papa_zooplankton.nc")
obs_papa
```

Merge them in a single dataset


```python
final_product = xr.merge([obs_bats, obs_hot, obs_papa])
final_product.attrs = {
    "stations": "BAT, HOT and PAPA",
    "creation date": datetime.now().isoformat(),
}
final_product
```

```python
final_product.to_netcdf("./final_zooplankton.nc")
```
